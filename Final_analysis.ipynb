### Now I will create a function to rank people based on how much calories per minute they have burned

```python
def rank_by_perfomance(Calories_per_minute):
    if Calories_per_minute <=4:
        return 'Needs Improvement'
    elif Calories_per_minute >=4.1 and Calories_per_minute <= 7:
        return 'Average'
    elif Calories_per_minute > 7.1 and Calories_per_minute <= 10:
        return 'Excellent'
    else:
        return 'Work'
    
df['Performance_Ranking'] = df['Calories_per_minute'].apply(rank_by_perfomance)
print(df) # This will show the whole Data Frame without truncating it.
```

### Now I will make a new variable to store the data of people with average perfomace in the gym and what are there numbers or count.

```python
average_performaner_list = df[df['Performance_Ranking'] == 'Average']
print(average_performaner_list.value_counts())
print(f"There are {len(average_performaner_list)} people in the gym with average performance.")
```
